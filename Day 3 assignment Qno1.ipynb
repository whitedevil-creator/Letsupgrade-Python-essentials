```
#using for loop all the prime no from 1 to 200
 
 
lower = 1
upper = 200
 
 
print("Prime numbers between",lower,"and",upper,"are:")
 
for a in range(lower,upper + 1):
 
   if a > 1:
       for i in range(2,a):
           if (a % i) == 0:
               break
       else:
           print(a)
```
